```python
import pandas as pd
import np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16,8)
```

```python
#sns.distplot([5,5,5,6,6,7])
#sns.histplot([5,5,5,6,6,7])
x = np.linspace(0,10*np.pi, 100)
y = np.sin(x)
plt.scatter(x, y)
```

```python
sns.histplot(y)
```

```python
sns.kdeplot(y)
```

```python
sns.distplot(y)
```

```python
sns.ecdfplot(y)
```

```python
sns.kdeplot(y, cumulative=True)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x[:, np.newaxis], y)
print(model.coef_, model.intercept_)
```

```python
pred = x * model.coef_[0] + model.intercept_
plt.plot(x, pred)
plt.scatter(x, y)
```

```python
sns.residplot(x,y)
```

```python
def maclaurin(x, y, inf=100):
    return (np.polyfit(x, y, inf)[:, np.newaxis] * (x ** np.arange(inf, -1, -1)[:, np.newaxis])).sum(axis=0)

plt.plot(x, maclaurin(x, y, inf=100))
plt.scatter(x, y)
```

```python
x = np.linspace(-10, 10, 100)
z = 1 / (1 + np.exp(-x))
plt.scatter(x, z)
```

```python
plt.scatter(x, z)
plt.plot(x, maclaurin(x, z))
plt.title("Graph of Sigmoid fitted using Maclaurin")
```
